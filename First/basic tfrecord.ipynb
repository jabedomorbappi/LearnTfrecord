```python
import tensorflow as tf

# Example data: a list of integers
data = [10, 20, 30]

# Create an Int64List
int64_list = tf.train.Int64List(value=data)

# Create a Feature using Int64List
feature = tf.train.Feature(int64_list=int64_list)


print(int64_list)
print(feature)
```

```python
def create_example(feature_data):
    # Create an Int64List feature
    int_feature = tf.train.Feature(int64_list=tf.train.Int64List(value=feature_data))

    # Create a Features object
    features = tf.train.Features(feature={'integer_feature': int_feature})

    # Create an Example object
    example = tf.train.Example(features=features)
    return example

# Example usage
example = create_example([1, 2, 3])
print(example)
```

##Multile Data with multi type

```python
int_list = tf.train.Int64List(value = [5, 6, 82])
float_list = tf.train.FloatList(value=[2.04, 4.08])
bytes_list = tf.train.BytesList(value=[str.encode('hi'), str.encode('hello')])
another_bytelist=tf.train.BytesList(value=[b'hi',b'hello'])
```

```python
feature1=tf.train.Feature(int64_list=int_list)
feature2=tf.train.Feature(float_list=float_list)
feature3=tf.train.Feature(bytes_list=bytes_list)
print(feature1)
print(feature2)
print(feature3)
```

```python
features=tf.train.Features(feature={'int_feature':feature1,'float_feature':feature2,'bytes_feature':feature3})
print(features)
```

```python
example=tf.train.Example(features=features)
print(example)
```

```python
def create_feature(int_list:tf.train.Int64List,float_list:tf.train.FloatList,bytes_list:tf.train.BytesList):

  return tf.train.Feature(int64_list=int_list,float_list=float_list,bytes_list=bytes_list)
create_feature(int_list,float_list,bytes_list)
```

#Here, I’m creating an instance of tf.train.Feature by passing all the lists created above. But when we print it, it only shows the byte list, no int list or float list are printed. The reason is tf.train.Feature has them as oneof fields. So only the list which was put at the last will be there, others will be cleared off.

```python
def create_feature(float_list:tf.train.FloatList,bytes_list:tf.train.BytesList,int_list:tf.train.Int64List):

  return tf.train.Feature(float_list=float_list,bytes_list=bytes_list,int64_list=int_list)
create_feature(float_list,bytes_list,int_list)
```
